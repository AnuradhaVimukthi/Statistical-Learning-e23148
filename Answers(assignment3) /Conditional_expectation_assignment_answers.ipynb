#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

7. Verify the identity:

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? 10. Using the joint distribution of $(X, Y)$ derived from the sample space:

# Q1) Prediction of Fraudulent Transactions

### 1. The Probability Space $(\Omega, \mathfrak{F}, P)$
The probability space consists of three components:
* **Sample Space ($\Omega$):** $\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\}$
* **$\sigma$-algebra ($\mathfrak{F}$):** $\mathfrak{F}=\mathfrak{P}(\Omega)$ (the power set, containing all subsets of $\Omega$)
* **Probability Measure ($P$):** Defined by the given atomic probabilities:
  $$P(\{\omega_1\})=0.50, \quad P(\{\omega_2\})=0.10, \quad P(\{\omega_3\})=0.20, \quad P(\{\omega_4\})=0.20$$
Since $0.50 + 0.10 + 0.20 + 0.20 = 1$, this forms a valid probability measure.

---

### 2. The $\sigma$-algebra generated by $Y$, $\sigma(Y)$
The random variable $Y$ maps outcomes to values based on the customer's risk class:
* $Y^{-1}(0) = \{\omega_1, \omega_2\}$
* $Y^{-1}(1) = \{\omega_3, \omega_4\}$

The $\sigma$-algebra $\sigma(Y)$ is the smallest collection of sets containing these preimages, closed under complements and unions:
$$\sigma(Y) = \left\{ \varnothing, \Omega, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\} \right\}$$

---

### 3. Measurability of $E[X \mid Y]$
By definition, $E[X \mid Y]$ must be $\sigma(Y)$-measurable. This means it can depend only on the value of $Y$, not on the full outcome $\omega$. That is, $E[X \mid Y] = g(Y)$ for some measurable map $g$.

Since $\sigma(Y)$ can distinguish only the two sets $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$, the random variable $E[X \mid Y]$ must be constant on each of these sets. We can write:
$$E[X \mid Y](\omega) = \begin{cases} a, & \omega \in \{\omega_1, \omega_2\} \\ b, & \omega \in \{\omega_3, \omega_4\} \end{cases}$$

---

### 4. Explicit Computation of $E[X \mid Y]$
We find $a$ and $b$ using the defining property of conditional expectation:
**On the set $\{\omega_1, \omega_2\}$ (where $Y=0$):**
$$\int_{\{\omega_1, \omega_2\}} E[X \mid Y] \, dP = \int_{\{\omega_1, \omega_2\}} X \, dP$$
$$a \cdot P(\{\omega_1, \omega_2\}) = (0 \cdot 0.50) + (1 \cdot 0.10)$$
$$a(0.60) = 0.10 \implies a = \frac{1}{6}$$

**On the set $\{\omega_3, \omega_4\}$ (where $Y=1$):**
$$\int_{\{\omega_3, \omega_4\}} E[X \mid Y] \, dP = \int_{\{\omega_3, \omega_4\}} X \, dP$$
$$b \cdot P(\{\omega_3, \omega_4\}) = (0 \cdot 0.20) + (1 \cdot 0.20)$$
$$b(0.40) = 0.20 \implies b = \frac{1}{2}$$

Equivalently, as an indicator function of $Y$:
$$E[X \mid Y] = \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}$$

---

### 5. Verification of the Defining Property
We must check that $\int_A E[X \mid Y] \, dP = \int_A X \, dP$ for every $A \in \sigma(Y)$:
* **$A = \varnothing$**: Both integrals are clearly $0$.
* **$A = \{\omega_1, \omega_2\}$**: $\frac{1}{6}(0.60) = 0.10$. The RHS yields $0(0.50) + 1(0.10) = 0.10$. $(\checkmark)$
* **$A = \{\omega_3, \omega_4\}$**: $\frac{1}{2}(0.40) = 0.20$. The RHS yields $0(0.20) + 1(0.20) = 0.20$. $(\checkmark)$
* **$A = \Omega$**: $\frac{1}{6}(0.60) + \frac{1}{2}(0.40) = 0.30$. The RHS yields $1(0.10) + 1(0.20) = 0.30$. $(\checkmark)$

---

### 6. Conditional PMF $p_{X|Y}(x \mid y)$
Using the product rule $p_{X|Y}(x \mid y) = \frac{P(X=x, Y=y)}{P(Y=y)}$, we calculate the conditional probability of fraud ($X=1$) for each class:
* **For $Y=0$:** $p_{X|Y}(1 \mid 0) = \frac{P(\{\omega_2\})}{0.60} = \frac{0.10}{0.60} = \mathbf{\frac{1}{6}}$ (Note: $p_{X|Y}(0 \mid 0) = \frac{5}{6}$)
* **For $Y=1$:** $p_{X|Y}(1 \mid 1) = \frac{P(\{\omega_4\})}{0.40} = \frac{0.20}{0.40} = \mathbf{\frac{1}{2}}$ (Note: $p_{X|Y}(0 \mid 1) = \frac{1}{2}$)

---

### 7. Verification of the Discrete Identity
We apply the discrete summary formula $\mathbb{E}[X \mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$:
* **Evaluation at $y=0$:** $(0 \cdot 5/6) + (1 \cdot 1/6) = \mathbf{\frac{1}{6}}$
* **Evaluation at $y=1$:** $(0 \cdot 1/2) + (1 \cdot 1/2) = \mathbf{\frac{1}{2}}$
These exactly match the measure-theoretic derivations from part 4.

---

### 8. Mean Square Error (MSE)
The optimal MMSE is the expected conditional variance $\mathbb{E}[\mathrm{Var}(X \mid Y)]$.
First, find the conditional variance for each risk class $y \in \{0, 1\}$. Since $X$ is a Bernoulli indicator, variance is $p(1-p)$:
$$\mathrm{Var}(X \mid Y=0) = \frac{1}{6}\left(1 - \frac{1}{6}\right) = \frac{5}{36}$$
$$\mathrm{Var}(X \mid Y=1) = \frac{1}{2}\left(1 - \frac{1}{2}\right) = \frac{1}{4} = \frac{9}{36}$$
Taking the expectation across $Y$:
$$\mathbb{E}[\mathrm{Var}(X \mid Y)] = \left(\frac{5}{36} \cdot 0.60\right) + \left(\frac{9}{36} \cdot 0.40\right) = \frac{1}{12} + \frac{1}{10} = \mathbf{\frac{11}{60} \approx 0.1833}$$

---

### 9. Global Estimate and the Law of Total Variance
If the bank uses the global mean $\mathbb{E}[X] = 0.30$, the error is the total variance $\mathrm{Var}(X)$:
$$\mathrm{Var}(X) = p(1-p) = 0.30(0.70) = \mathbf{0.21}$$

The **Law of Total Variance** decomposes this into:
$$\mathrm{Var}(X) = \mathbb{E}[\mathrm{Var}(X \mid Y)] + \mathrm{Var}(\mathbb{E}[X \mid Y])$$
* **Unexplained Variance (Optimal Error):** $\mathbb{E}[\mathrm{Var}(X \mid Y)] = \mathbf{0.1833}$
* **Explained Variance (Error Reduction):** $$\mathbb{E}[(\mathbb{E}[X \mid Y])^2] = \left(\frac{1}{36} \cdot 0.60\right) + \left(\frac{1}{4} \cdot 0.40\right) = 0.0166 + 0.1 = 0.1166$$
  $$\mathrm{Var}(\mathbb{E}[X \mid Y]) = 0.1166 - (0.3)^2 = \mathbf{0.0267}$$

Summary Check: $0.21 (\text{Total Error}) = 0.1833 (\text{Remaining Error}) + 0.0267 (\text{Error Reduction})$.

---

### 10. Information Theory Quantities
* **a) Entropy $H(X)$:**
  $$H(X) = -(0.70 \log_2 0.70 + 0.30 \log_2 0.30) \approx \mathbf{0.881 \text{ bits}}$$
* **b) Conditional Entropy $H(X \mid Y)$:**
  $$H(X \mid Y=0) = -\left( \frac{1}{6} \log_2 \frac{1}{6} + \frac{5}{6} \log_2 \frac{5}{6} \right) \approx 0.650 \text{ bits}$$
  $$H(X \mid Y=1) = -\left( \frac{1}{2} \log_2 \frac{1}{2} + \frac{1}{2} \log_2 \frac{1}{2} \right) = 1.00 \text{ bit}$$
  $$H(X \mid Y) = (0.60 \times 0.650) + (0.40 \times 1.00) = \mathbf{0.790 \text{ bits}}$$
* **c) Mutual Information $I(X; Y)$:**
  $$I(X; Y) = H(X) - H(X \mid Y) = 0.881 - 0.790 = \mathbf{0.091 \text{ bits}}$$
* **d) Uncertainty Removed:**
  $$\text{Percentage} = \frac{I(X;Y)}{H(X)} \times 100\% = \frac{0.091}{0.881} \times 100\% \approx \mathbf{10.3\%}$$

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

# Q2) Conditional Density of the Multivariate Gaussian

To derive the conditional distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the definition of conditional probability for densities:
$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

### Step 1: Decomposition of the Joint Quadratic Form
The joint density $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})$ relies on a quadratic form $Q_{joint}$. Using the Schur complement of $\boldsymbol{\Sigma}_{YY}$, denoted as $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$, we can factor $Q_{joint}$ completely:
$$Q_{joint} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y) + (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$
Where the shifted mean is:
$$\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$$

### Step 2: Subtraction of the Marginal Exponent
The marginal density $f_{\mathbf{Y}}(\mathbf{y})$ has the quadratic exponent:
$$Q_{marginal} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y)$$

When we divide the densities, we subtract the exponents ($e^{-Q_{joint}/2} / e^{-Q_{marginal}/2} = e^{-(Q_{joint}-Q_{marginal})/2}$). The terms involving purely $\mathbf{y}$ cancel out, leaving:
$$Q_{conditional} = Q_{joint} - Q_{marginal} = (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

### Step 3: Simplification of the Normalization Constants
We now divide the normalization constants of the joint and marginal distributions:
$$\frac{\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}}{\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}} = \frac{1}{\sqrt{(2\pi)^n \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}$$

Using the determinant identity for block matrices, $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$. Therefore, the term $\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}$ simplifies exactly to $|\mathbf{S}_{XX}|$.

### Step 4: Final Conditional Density
Combining the simplified constant with the remaining quadratic form yields the final conditional density:
$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**

2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

Identify the distribution of the random varaible $(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.

Calculate $I(X; Y) = h(X) - h(X \mid Y)$.

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*

## Answer

# Q3) Continuous Fault Monitoring of a Spindle Motor

### 1. Joint Distribution of $(X, Y)$
Since $X$ and $Q$ are independent normal random variables, their sum $Y = X + Q$ is also normally distributed.
* **Mean:** $\mathbb{E}[Y] = \mathbb{E}[X] + \mathbb{E}[Q] = 0 + 0 = 0$
* **Variance:** Since $X \perp Q$, $\mathrm{Var}(Y) = \mathrm{Var}(X) + \mathrm{Var}(Q) = 1 + 4 = 5$
* **Covariance:** $\text{Cov}(X, Y) = \mathbb{E}[XY] - \mathbb{E}[X]\mathbb{E}[Y] = \mathbb{E}[X(X+Q)] = \mathbb{E}[X^2] + \mathbb{E}[XQ]$. Since $\mathbb{E}[X^2] = 1$ and $\mathbb{E}[XQ] = 0$, we get $\text{Cov}(X, Y) = 1$.

The joint distribution is:
$$\begin{pmatrix} X \\ Y \end{pmatrix} \sim \mathscr{N} \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & 1 \\ 1 & 5 \end{pmatrix} \right)$$

---

### 2. Information Geometry
$X$ is not $\sigma(Y)$-measurable because there is no deterministic function $g$ such that $X = g(Y)$. For any observed $Y=y$, there are infinitely many valid physical pairings of $(x, q)$ such that $x+q=y$.

**Practically:** A high power reading ($Y=5$) is ambiguous. The sensor cannot distinguish whether this is caused by a true "Thermal Fault" (high $X$) or if the machine is simply cutting a "Hard Material" (high $Q$).

---

### 3. Conditional Expectation (The Best Estimate)
For bivariate normal variables, the optimal mean-square estimate is:
$$\mathbb{E}[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\mathrm{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0)$$
$$\mathbf{\mathbb{E}[X \mid Y] = 0.2Y}$$
**Interpretation:** Because the mechanical noise $Q$ ($\sigma^2=4$) is four times more volatile than the thermal signal $X$ ($\sigma^2=1$), the model only "trusts" 20% of the total power fluctuation as being driven by temperature.

---

### 4. Verification of the Residual
Let $\widehat{X} = 0.2Y$. The residual is $Z = X - 0.2Y$.
* **Mean:** $\mathbb{E}[Z] = \mathbb{E}[X] - 0.2\mathbb{E}[Y] = 0 - 0 = \mathbf{0}$
* **Orthogonality:** $\text{Cov}(Z, Y) = \mathbb{E}[ZY] = \mathbb{E}[(X - 0.2Y)Y] = \mathbb{E}[XY] - 0.2\mathbb{E}[Y^2]$.
  Using $\mathbb{E}[XY] = 1$ and $\mathbb{E}[Y^2] = 5$:
  $$\text{Cov}(Z, Y) = 1 - 0.2(5) = 1 - 1 = \mathbf{0}$$
The error is uncorrelated with our observation, meaning all linear information has been fully extracted from $Y$.

---

### 5. Differential Entropy
Recall $h(\mathscr{N}(\mu, \sigma^2)) = \frac{1}{2}\ln(2\pi e \sigma^2)$.
* **Marginal $h(X)$:** With $\sigma_X^2 = 1$:
  $$h(X) = \frac{1}{2}\ln(2\pi e) \approx \mathbf{1.4189 \text{ nats}}$$
* **Conditional $h(X \mid Y)$:** The conditional variance is $\sigma_{X \mid Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\mathrm{Var}(Y)} = 1 - \frac{1}{5} = 0.8$.
  $$h(X \mid Y) = \frac{1}{2}\ln(2\pi e \cdot 0.8) \approx \mathbf{1.3073 \text{ nats}}$$

---

### 6. Mutual Information
$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2}\ln(1) - \frac{1}{2}\ln(0.8) = \frac{1}{2}\ln\left(\frac{1}{0.8}\right) = \frac{1}{2}\ln(1.25) \approx \mathbf{0.1116 \text{ nats}}$$

**Percentage Uncertainty Removed:** Measured by variance reduction, we went from a variance of $1.0$ to $0.8$, a **$20\%$ reduction**.
If the mechanical noise variance increased to $\sigma_Q^2 = 100$, the conditional variance becomes $1 - \frac{1}{101} \approx 0.9901$. The mutual information $I(X;Y)$ would approach $0$, and only $\mathbf{\sim 0.99\%}$ of the uncertainty would be removed.

---

### 7. Diagnostic Interpretation
The manager's suggestion to use $X = Y$ acts as a **Naive Estimator**.
* If we use $X = Y$, our error variance is $\mathbb{E}[(X-Y)^2] = \mathbb{E}[Q^2] = 4$.
* By using the conditional expectation $\mathbb{E}[X \mid Y] = 0.2Y$, our error variance drops to $0.8$.

**Shrinkage:** The coefficient $0.2$ is a shrinkage estimator. It proves that when dealing with a heavily noisy sensor, the mathematically safest bet is to "shrink" the reading back toward the historical average ($0$). This conservatively assumes that extreme spikes in power are overwhelmingly due to mechanical load rather than instant catastrophic thermal failure.
